### 讀取範例影片

```python
import glob
video_path = glob.glob('*Test*.mp4')[0]
import cv2
cap = cv2.VideoCapture(video_path)
seconds = cap.get(cv2.CAP_PROP_FRAME_COUNT) / cap.get(cv2.CAP_PROP_FPS)
frames = []
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    frames.append(frame)
assert len(frames) == cap.get(cv2.CAP_PROP_FRAME_COUNT)
```

### 播放範例影片

```python
from IPython.display import HTML
from base64 import b64encode
with open(video_path, "rb") as f:
    data_url = "data:video/mp4;base64," + b64encode(f.read()).decode()
HTML(f'<video width=1280 controls><source src="{data_url}" type="video/mp4"></video>')
```

```python
#@title 設定區間及速度倍率後，執行下方程式格，可重複實測
開始秒數 =  3.5#@param {type:"number"}
結束秒數 =  4.5#@param {type:"number"}
速度倍率 = 0.05 #@param {type:"number"}
if not 0 <= 開始秒數 < seconds:
    print('開始秒數有誤！')
if not 0 <= 結束秒數 < seconds:
    print('結束秒數有誤！')
```

### 製作慢速影片

```python
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
fps = cap.get(cv2.CAP_PROP_FPS)
frame_size = (
    int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)),
    int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)))
out = cv2.VideoWriter('out.mp4', fourcc, int(fps * 速度倍率), frame_size, 1)
import numpy as np
begin = int(cap.get(cv2.CAP_PROP_FPS) * 開始秒數)
end   = int(cap.get(cv2.CAP_PROP_FPS) * 結束秒數)
for frame in frames[begin:end]:
    out.write(frame)
out.release()
!ffmpeg -i out.mp4 -acodec libx264 avc.mp4 -y
```
